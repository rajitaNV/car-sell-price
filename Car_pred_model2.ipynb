```python
import pandas as pd
```

```python
df=pd.read_csv('car data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
print(df['Seller_Type'].unique())
print(df['Fuel_Type'].unique())
print(df['Transmission'].unique())
print(df['Owner'].unique())
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
final_df=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_df.head()
```

```python
final_df['Curr_year']=2021
```

```python
final_df.head()
```

```python
final_df['No_of_yrs']=final_df['Curr_year']-final_df['Year']
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df.drop(['Year'],axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.drop(['Curr_year'],axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df=pd.get_dummies(final_df,drop_first=True)
```

```python
final_df.head()
```

```python
final_df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.pairplot(final_df)
```

```python
cormat=final_df.corr()
top_corr_features=cormat.index
plt.figure(figsize=(20,20))
g=sns.heatmap(final_df[top_corr_features].corr(),annot=True,cmap='RdYlGn')
```

```python
x=final_df.iloc[:,1:]
y=final_df.iloc[:,0]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
feat_imp=pd.Series(model.feature_importances_,index=x.columns)
feat_imp.nlargest(5).plot(kind='barh')
plt.show
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
X_train.shape
```

```python
# Model Building using Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
linearRegression = LinearRegression()
linearRegression.fit(X_train, y_train)
y_pred_lr = linearRegression.predict(X_test)
r2_score(y_test, y_pred_lr)
```

```python
sns.distplot(y_test-y_pred_lr)
```

```python
# Model Building using RandomForestRegressor
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 100)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
r2_score(y_test, y_pred_rf)
```

```python
sns.distplot(y_test-y_pred_rf)
```

```python
plt.scatter(y_test,y_pred_rf)
```

```python
import pickle
# open a file, where you ant to store the data
file = open('random_forest_regression_model2.pkl', 'wb')

# dump information to that file
pickle.dump(rf, file)
file.close()
```
